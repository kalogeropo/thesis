```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Sample sizes
sample_1 = 250
sample_2 = 250

# Intersection
sample_intersection = 75

# Create the Venn diagram
venn = venn2(subsets=(sample_1, sample_2, sample_intersection), set_labels = (None, None))

# Add labels
# venn.get_label_by_id('00').set_text('TN')
venn.get_label_by_id('10').set_text('FN\nR: Σχετικά έγγραφα')
venn.get_patch_by_id('10').set_color("#9DD5FB")
venn.get_label_by_id('01').set_text('FP\nA: Έγγραφα απάντησης')
venn.get_patch_by_id('01').set_color("#F4E637")
venn.get_label_by_id('11').set_text('TP')
venn.get_patch_by_id('11').set_color('#C2DCA7')
# Create an outer circle as Universe
fig = plt.gcf()
ax = fig.gca()
circle = patches.Circle((0.5, 0.5), alpha=0.05, color='gray')
ax.add_patch(circle)

# Add label for Universe
plt.text(0., .47, 'TN', horizontalalignment='center', verticalalignment='center', fontsize=10)
plt.tight_layout()
plt.show()
```
